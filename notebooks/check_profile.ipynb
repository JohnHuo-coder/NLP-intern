```python
import sys
sys.path.append("..")
from scripts.text_cleaning import TextCleaner
import pandas as pd
# Use this to guide your cleaning strategy:
df = pd.read_csv("../data/processed/listing_sample.csv")
cleaner = TextCleaner()
profile = cleaner.profile_column(df, 'remarks')
common_terms = profile["common_terms"]
common_terms = [item['term'] for item in common_terms]
print("Null rate: ", profile["null_rate"])
print("Average remark length in character: ", profile["avg_length"])
print("Average remark length in words: ", profile["avg_num_words"])
print("Common terms in remarks: ", common_terms)
print("Number of times mentioning price: ", profile["price_mentions"])
print(f"Common abbreviations: {profile['common_abbreviations']}")
print(f"{profile['has_html']} HTML components found")
print("Common html in remarks: ", profile["html_examples"])
print(f"{profile['has_unicode']} unicode found")
print("Common unicode in remarks: ", profile["unicode_examples"])
print(f"{profile['price_mentions']} price term found")
print("Common price in remarks: ", profile["price_counter"])
print(f"{profile['measurement_mentions']} measurement found")
print("Common measurement in remarks: ", profile["measurement_counter"])
```

```python
df_cleaned = df.copy()
comparison_df = cleaner.sample_compare(df_cleaned["remarks"])
df_cleaned["remarks"] = cleaner.clean_column(df_cleaned["remarks"])
profile = cleaner.profile_column(df_cleaned, 'remarks')
common_terms = profile["common_terms"]
common_terms = [item['term'] for item in common_terms]
print("Null rate: ", profile["null_rate"])
print("Average remark length in character: ", profile["avg_length"])
print("Average remark length in words: ", profile["avg_num_words"])
print("Common terms in remarks: ", common_terms)
print("Number of times mentioning price: ", profile["price_mentions"])
print(f"Common abbreviations: {profile['common_abbreviations']}")
print(f"{profile['has_html']} HTML components found")
print("Common html in remarks: ", profile["html_examples"])
print(f"{profile['has_unicode']} unicode found")
print("Common unicode in remarks: ", profile["unicode_examples"])
print(f"{profile['price_mentions']} price term found")
print("Common price in remarks: ", profile["price_counter"])
print(f"{profile['measurement_mentions']} measurement found")
print("Common measurement in remarks: ", profile["measurement_counter"])
df_cleaned.to_csv("../data/processed/listing_sample_cleaned.csv")
comparison_df.to_csv("../data/processed/cleaned_sample_comparison.csv")
comparison_df.to_html("../data/processed/cleaned_sample_comparison.html", escape=False)
```
